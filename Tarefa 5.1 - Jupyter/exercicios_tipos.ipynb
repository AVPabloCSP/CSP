### Exercicio 2: Conversións de Tipo

A partires dos seguintes elementos:

```python
numero_str = "1234"
numero_float = 45.67
lista_numeros = [1, 2, 3, 4, 5]
```

Realiza as seguintes conversións:

1. numero_str a enteiro
2. numero_float a enteiro (qué pasa coa parte decimal?)
3. lista_numeros a tupla e a conxunto
4. Crea un string que combine numero_str e numero_float

```python
# Escribe aquí o código da túa solución
numero_str = "1234"
numero_float = 45.67
lista_numeros = [1, 2, 3, 4, 5]

# 1. String a enteiro
entero = int(numero_str)
print("String a enteiro:", entero)

# 2. Float a enteiro
entero_float = int(numero_float)
print("Float a enteiro:", entero_float)
# 45 → Ao converter a enteiro, Python **trunca** a parte decimal, non fai arredondamento.
# 3. Lista a tupla e a conxunto
tupla = tuple(lista_numeros)
conxunto = set(lista_numeros)

print("Lista a tupla:", tupla)
print("Lista a conxunto:", conxunto)

# 4. Combinar string e float nun string
combinado = numero_str + " " + str(numero_float)
print("String combinado:", combinado)
```

### Exercicio 3: Cadeas

A partires desta cadea:

```python
texto = "  Python é unha linguaxe de programación super famosa  "
```

Realiza estas operacións:

1. Elimina espazos ao comezo e ao final
2. Convirte a maiúsculas
3. Reempraza "linguaxe" por "lang"
4. Divide o string en palabras
5. Atopa a posición de "programación"

```python
# Escribe aquí o código da túa solución
texto = "  Python é unha linguaxe de programación super famosa  "

# 1. Eliminar espazos ao comezo e ao final
texto_limpo = texto.strip()
print("Texto sen espazos:", texto_limpo)

# 2. Converter a maiúsculas
texto_maiusculas = texto_limpo.upper()
print("En maiúsculas:", texto_maiusculas)

# 3. Reemprazar 'linguaxe' por 'lang'
texto_reemprazado = texto_limpo.replace("linguaxe", "lang")
print("Texto reemprazado:", texto_reemprazado)

# 4. Dividir o string en palabras
palabras = texto_limpo.split()
print("Palabras:", palabras)

# 5. Atopar a posición de 'programación'
posicion = texto_limpo.find("programación")
print("Posición de 'programación':", posicion)
```

### Exercicio 4: Listas

A partires da seguinte lista:

```python
frutas = ["mazá", "plátano", "laranxa", "uva", "mazá"]
```

Realiza estas operacións:

1. Engade "pera" ao final
2. Inserta "kiwi" na posición 2
3. Elimina a primera "mazá"
4. Ordena a lista alfabéticamente
5. Crea unha nova lista coas frutas que teñen máis de 5 letras


```python
# Escribe aquí o código da túa solución
frutas = ["mazá", "plátano", "laranxa", "uva", "mazá"]

# 1. Engadir "pera" ao final
frutas.append("pera")

# 2. Inserir "kiwi" na posición 2
frutas.insert(2, "kiwi")

# 3. Eliminar a primeira "mazá"
frutas.remove("mazá")

# 4. Ordenar a lista alfabeticamente
frutas.sort()

print("Lista ordenada:", frutas)

# 5. Nova lista con frutas de máis de 5 letras
frutas_longas = []

for froita in frutas:
    if len(froita) > 5:
        frutas_longas.append(froita)

print("Frutas con máis de 5 letras:", frutas_longas)
```

### Exercicio 7: Lista de Diccionarios

Crea unha lista de diccionarios con información de estudantes:

```python
estudantes = [
    {"nome": "Ana", "idade": 20, "nota": 8.5},
    {"nome": "Carlos", "idade": 22, "nota": 7.8},
    {"nome": "María", "idade": 21, "nota": 9.2}
]
```

Realiza estas operacións:

1. Engade un nopvo estudante.
2. Atopa o estudiante con maior nota.
3. Calcula o promedio de idades.
4. Crea unha lista só cos nomes.
5. Ordena os estudantes por nota (descendente).

```python
# Escribe aquí o código da túa solución
estudantes = [
    {"nome": "Ana", "idade": 20, "nota": 8.5},
    {"nome": "Carlos", "idade": 22, "nota": 7.8},
    {"nome": "María", "idade": 21, "nota": 9.2}
]

# 1. Engadir un novo estudante
novo_estudante = {"nome": "Luis", "idade": 23, "nota": 8.9}
estudantes.append(novo_estudante)

# 2. Atopar o estudante con maior nota
maior_nota = max(estudantes, key=lambda e: e["nota"])
print("Estudante con maior nota:", maior_nota)

# 3. Calcular o promedio de idades
promedio_idade = sum(e["idade"] for e in estudantes) / len(estudantes)
print("Promedio de idades:", promedio_idade)

# 4. Crear unha lista só cos nomes
nombres = [e["nome"] for e in estudantes]
print("Lista de nomes:", nombres)

# 5. Ordenar os estudantes por nota (descendente)
estudantes_ordenados = sorted(estudantes, key=lambda e: e["nota"], reverse=True)
print("Estudantes ordenados por nota (descendente):", estudantes_ordenados)
```

### Exercicio 8: Conxuntos

A partir destes conxuntos:

```python
conx_a = {1, 2, 3, 4, 5}
conx_b = {4, 5, 6, 7, 8}
```

Realiza estas operacións:

1. Unión de ambos conxuntos
2. Intersección
3. Diferencia (A - B)
4. Diferencia simétrica
5. Verifica si 3 está en A

```python
# Escribe aquí o código da túa solución
conx_a = {1, 2, 3, 4, 5}
conx_b = {4, 5, 6, 7, 8}

# 1. Unión
union = conx_a | conx_b  # tamén se pode con conx_a.union(conx_b)
print("Unión:", union)

# 2. Intersección
interseccion = conx_a & conx_b  # tamén se pode con conx_a.intersection(conx_b)
print("Intersección:", interseccion)

# 3. Diferencia (A - B)
diferencia = conx_a - conx_b  # tamén se pode con conx_a.difference(conx_b)
print("Diferencia (A - B):", diferencia)

# 4. Diferencia simétrica
diferencia_simetrica = conx_a ^ conx_b  # tamén se pode con conx_a.symmetric_difference(conx_b)
print("Diferencia simétrica:", diferencia_simetrica)

# 5. Verificar se 3 está en A
esta_en_a = 3 in conx_a
print("Está 3 en A?", esta_en_a)
```

### Exercicio 9: Programa Integrador

Crea un programa que:

1. Pida ao usuario o seu nome, idade e cidade de residencia
2. Almacene esa información nun diccionario
3. Convirta a idade a enteiro
4. Cree unha lista de hobbies (pregunta por 3 polo menos)
5. Amose toda a información formateada de xeito lexible

```python
# Escribe aquí o código da túa solución
# 1. Pedir datos ao usuario
nome = input("Introduce o teu nome: ")
idade = input("Introduce a túa idade: ")
cidade = input("Introduce a túa cidade de residencia: ")

# 2. Almacenar nun diccionario
usuario = {
    "nome": nome,
    "idade": int(idade),  # 3. Convertir a idade a enteiro
    "cidade": cidade
}

# 4. Crear lista de hobbies
hobbies = []
print("Introduce polo menos 3 hobbies:")
for i in range(3):
    hobby = input(f"Hobby {i+1}: ")
    hobbies.append(hobby)

# Podes engadir máis hobbies se queres
while True:
    resposta = input("Queres engadir outro hobby? (s/n): ").lower()
    if resposta == "s":
        hobby = input("Novo hobby: ")
        hobbies.append(hobby)
    else:
        break

usuario["hobbies"] = hobbies

# 5. Amosar toda a información de xeito lexible
print("\n--- Información do usuario ---")
print(f"Nome: {usuario['nome']}")
print(f"Idade: {usuario['idade']} anos")
print(f"Cidade: {usuario['cidade']}")
print("Hobbies:")
for h in usuario["hobbies"]:
    print(f" - {h}")
```

### Exercicio 10: Combinación de Tipos Complexos

Temos un Sistema de biblioteca:

```python
biblioteca = {
    "libros": [
        {
            "titulo": "Cien años de soledad",
            "autor": "Gabriel García Márquez", 
            "ano": 1967,
            "xeneros": {"Realismo mágico", "Novela"},
            "disponible": True
        },
        {
            "titulo": "1984",
            "autor": "George Orwell",
            "ano": 1949,
            "xeneros": {"Ciencia ficción", "Distopía"},
            "disponible": False
        }
    ],
    "usuarios": {
        "001": {"nome": "Ana", "libros_prestados": ["1984"]},
        "002": {"nome": "Carlos", "libros_prestados": []}
    }
}
```

Realiza estas operacións:

1. Engade un novo libro.
2. Marca un libro como non dispoñible.
3. Atopa todos os libros dun xénero.
4. Libros dispoñibles por autor.
5. Usuario con máis libros prestados.

```python
# Escribe aquí o código da túa solución
biblioteca = {
    "libros": [
        {
            "titulo": "Cien años de soledad",
            "autor": "Gabriel García Márquez", 
            "ano": 1967,
            "xeneros": {"Realismo mágico", "Novela"},
            "disponible": True
        },
        {
            "titulo": "1984",
            "autor": "George Orwell",
            "ano": 1949,
            "xeneros": {"Ciencia ficción", "Distopía"},
            "disponible": False
        }
    ],
    "usuarios": {
        "001": {"nome": "Ana", "libros_prestados": ["1984"]},
        "002": {"nome": "Carlos", "libros_prestados": []}
    }
}

# 1. Engadir un novo libro
novo_libro = {
    "titulo": "El Principito",
    "autor": "Antoine de Saint-Exupéry",
    "ano": 1943,
    "xeneros": {"Infantil", "Novela"},
    "disponible": True
}
biblioteca["libros"].append(novo_libro)

# 2. Marcar un libro como non dispoñible (por exemplo, "Cien años de soledad")
for libro in biblioteca["libros"]:
    if libro["titulo"] == "Cien años de soledad":
        libro["disponible"] = False
        break

# 3. Atopar todos os libros dun xénero (por exemplo, "Novela")
xenero_buscado = "Novela"
libros_por_xenero = [libro["titulo"] for libro in biblioteca["libros"] if xenero_buscado in libro["xeneros"]]
print("Libros do xénero 'Novela':", libros_por_xenero)

# 4. Libros dispoñibles por autor (diccionario: autor -> lista de títulos dispoñibles)
libros_disponibles_por_autor = {}
for libro in biblioteca["libros"]:
    if libro["disponible"]:
        autor = libro["autor"]
        libros_disponibles_por_autor.setdefault(autor, []).append(libro["titulo"])

print("Libros dispoñibles por autor:", libros_disponibles_por_autor)

# 5. Usuario con máis libros prestados
usuario_max = max(biblioteca["usuarios"].items(), key=lambda item: len(item[1]["libros_prestados"]))
id_usuario, datos_usuario = usuario_max
print("Usuario con máis libros prestados:", datos_usuario["nome"], "-", len(datos_usuario["libros_prestados"]), "libros")
```

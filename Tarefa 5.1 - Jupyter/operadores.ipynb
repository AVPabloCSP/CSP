### Exercicio 1: Operadores Aritméticos

Calcula a área e perímetro dun rectángulo:
- Pide ao usuario base e altura.
- Calcula área.
- Calcula perímetro.
- Utiliza `//` e `%`.
- Amosa os resultados utilizando diferentes formatos de cadeas.

```python
# Pon a túa solución aquí
# Pedir datos ao usuario
base = int(input("Introduce a base do rectángulo en cm: "))
altura = int(input("Introduce a altura do rectángulo en cm: "))

# Calcular área
area = base * altura

# Calcular perímetro
perimetro = 2 * (base + altura)

# Exemplo de uso de // e %
# Imos dividir o perímetro en dúas partes para ver división entera e resto
perimetro_div = perimetro // 2
perimetro_rest = perimetro % 2

# Amosar resultados en diferentes formatos

# Usando concatenación de strings
print("Área do rectángulo: " + str(area) + " cm²")
print("Perímetro do rectángulo: " + str(perimetro) + " cm")

# Usando f-strings
print(f"Área (f-string): {area} cm²")
print(f"Perímetro (f-string): {perimetro} cm (metade: {perimetro_div}, resto: {perimetro_rest})")

# Usando format()
print("Área (format()): {} cm²".format(area))
print("Perímetro (format()): {} cm".format(perimetro))
```

### Exercicio 2: Operadores de Asignación

Simula unha conta bancaria:
- Saldo inicial: 1000
- Usa += para depositar 500
- Usa -= para retirar 300
- Usa *= para aplicar interés do 10%
- Usa /= para dividir entre 2 (metade para aforros)
- Amosa saldo despois de cada operación
- Utiliza diferentes formatos de cadeas.

```python
# Pon a túa solución aquí
# Saldo inicial
saldo = 1000
print("Saldo inicial:", saldo)

# Depositar 500 con +=
saldo += 500
print(f"Despois de depositar 500: {saldo}")  # f-string

# Retirar 300 con -=
saldo -= 300
print("Despois de retirar 300: " + str(saldo))  # concatenación clásica

# Aplicar interese do 10% con *=
saldo *= 1.10
print("Despois de aplicar interese do 10%: {:.2f}".format(saldo))  # format() con 2 decimais

# Dividir entre 2 con /=
saldo /= 2
print(f"Despois de dividir entre 2 (aforros): {saldo:.2f}")  # f-string con 2 decimais
```

### Exercicio 3: Operadores de Comparación

Sistema de verificación de acceso:
- Pide idade e se é socio ou non.
- Verifica se idade >= 18 e <= 65
- Verifica se idade <> 0
- Usa todos estes operadores: >, <, >=, <=, ==, !=
- O mesmo: utiliza diferentes formatos de cadeas para amosar os resultados.

```python
# Pon a túa solución aquí
# Pedir datos ao usuario
idade = int(input("Introduce a túa idade: "))
socio = input("Es socio? (si/no): ").lower() == "si"

# Comprobar idade >= 18 e <= 65
acceso_por_idade = idade >= 18 and idade <= 65
print(f"Idade entre 18 e 65? {acceso_por_idade}")  # f-string

# Comprobar idade > 18
maior_18 = idade > 18
print("Idade maior que 18? " + str(maior_18))  # concatenación clásica

# Comprobar idade < 65
menor_65 = idade < 65
print("Idade menor que 65? {}".format(menor_65))  # format()

# Comprobar idade == 0
idade_cero = idade == 0
print(f"Idade é 0? {idade_cero}")

# Comprobar idade != 0
idade_non_cero = idade != 0
print("Idade non é 0? " + str(idade_non_cero))

if socio:
    if acceso_por_idade:
        print("Pode acceder como socio")
    else:
        print("É socio pero non cumpre idade")
else:
    print("Non é socio")


```

### Exercicio 4: Operadores Lóxicos

Determina se un ano é bisesto:
- Divisible entre 4.
- Non divisible entre 100, excepto se é divisible entre 400.
- Non encadenes condicionais senón condicións.
- Exemplo: 2024 (bisesto), 1900 (non bisesto), 2000 (bisesto).

```python
# Pon a túa solución aquí
# Pedir o ano ao usuario
ano = int(input("Introduce un ano: "))

# Determinar se é bisesto
# Condicións:
divisible_4 = ano % 4 == 0
non_divisible_100 = ano % 100 != 0
divisible_400 = ano % 400 == 0

# Bisesto se é divisible por 4 e non por 100, ou divisible por 400
bisesto = (divisible_4 and non_divisible_100) or divisible_400

# Amosar resultado
print(f"É o ano {ano} bisesto? {bisesto}")
```

### Exercicio 5: Operadores de Identidade

Analiza diferenzas entre == e is:
- Crea dúas listas (lista1, lista2) iguais: [1, 2, 3]
- Crea unha variable (lista3) que apunte lista2, por exemplo.
- Explica resultados de comparar `==` e `is`.

```python
# Pon a túa solución aquí
# Crear listas
lista1 = [1, 2, 3]
lista2 = [1, 2, 3]

# Apuntar lista3 a lista2
lista3 = lista2

# Comparación de valores
print("lista1 == lista2:", lista1 == lista2)  # True, porque teñen os mesmos valores
print("lista2 == lista3:", lista2 == lista3)  # True, tamén teñen os mesmos valores

# Comparación de identidade
print("lista1 is lista2:", lista1 is lista2)  # False, son obxectos diferentes en memoria
print("lista2 is lista3:", lista2 is lista3)  # True, apuntan ao mesmo obxecto
```

### Exercicio 6: Operadores de Pertenza

Buscador de vogais en texto:
- Pide unha frase ao usuario.
- Usa `in` para verificar se contén cada vogal.
- Usa `not in` para verificar se non contén números.
- Conta cantas vogais diferentes ten.

```python
# Pon a túa solución aquí
# Pedir unha frase ao usuario
frase = input("Introduce unha frase: ").lower()  # Pasamos a minúsculas para facilitar a comprobación

vogais = "aeiou"
vogais_encontradas = set()  # Conxunto para almacenar só vogais únicas

# Comprobar cada vogal
for v in vogais:
    if v in frase:
        print(f"A vogal '{v}' está na frase")
        vogais_encontradas.add(v)
    else:
        print(f"A vogal '{v}' NON está na frase")

# Comprobar se hai números
numeros = "0123456789"
contén_numeros = any(n in frase for n in numeros)
print(f"A frase contén números? {contén_numeros}")

# Cantidade de vogais diferentes
print(f"Número de vogais diferentes na frase: {len(vogais_encontradas)}")
```

### Exercicio 7: Operadores de Cadeas

Manipulación de strings:
- Concatena o teu nome, apelido e o nome do módulo (CSP).
- Repite un patrón 3 veces.
- Verifica se "ola" está nunha frase.
- Extrae subcadeas con `slicing`.
- Converte a maiúsculas e a minúsculas.

```python
# Pon a túa solución aquí
# Concatena nome, apelido e módulo
nome = "Pablo"
apelido = "Antelo"
modulo = "CSP"

frase_completa = nome + " " + apelido + " - " + modulo
print("Concatenación:", frase_completa)

# Repetir un patrón 3 veces
patron = "-*-"
patron_repetido = patron * 3
print("Patrón repetido:", patron_repetido)

# Verificar se "ola" está nunha frase
frase = "Ola, benvido ao curso de Python"
contén_ola = "ola" in frase.lower()  # pasamos a minúsculas para non depender de maiúsculas
print("Contén 'ola'? ", contén_ola)

# Extraer subcadeas con slicing
subcadea = frase[0:3]  # primeiros 3 caracteres
print("Subcadea (primeiros 3):", subcadea)

subcadea_final = frase[-6:]  # últimos 6 caracteres
print("Subcadea (últimos 6):", subcadea_final)

# Maiúsculas e minúsculas
print("Maiúsculas:", frase.upper())
print("Minúsculas:", frase.lower())
```

### Exercicio 8: Walrus

Validador de contrasinal con walrus:
- Pide contrasinal ata que cumpra requisitos.
- Lonxitude >= 8 caracteres.
- Contén maiúscula e número.
- Usa walrus para non repetir len(contrasinal).

```python
# Pon a túa solución aquí
import re  # Para buscar maiúsculas e números

print("Introduce un contrasinal seguro:")

while True:
    # Pedimos contrasinal e gardamos lonxitude co walrus
    if (lonxitude := len(contrasinal := input("Contrasinal: "))) >= 8:
        # Verificar se contén maiúscula
        ten_maiuscula = bool(re.search(r"[A-Z]", contrasinal))
        # Verificar se contén número
        ten_numero = bool(re.search(r"\d", contrasinal))

        if ten_maiuscula and ten_numero:
            print(f"Contrasinal válido! (lonxitude {lonxitude})")
            break
        else:
            print("O contrasinal debe ter polo menos unha maiúscula e un número.")
    else:
        print(f"O contrasinal é demasiado curto ({lonxitude} caracteres). Mínimo 8.")
```

### Exercicio 9: Combinación de Operadores

Calculadora de IMC (xa a fixemos pero desta volta é máis completa):
- peso / altura²
- Clasifica segundo táboa OMS
- Baixo peso: < 18.5
- Normal: 18.5 - 24.9
- Sobrepeso: 25 - 29.9
- Obesidade: >= 30
- Usa operadores de comparación e lóxicos
- Opcional: walrus para evitar repetir cálculo

```python
# Pon a túa solución aquí
# Pedir datos
peso = float(input("Introduce o teu peso en kg: ")) # Neste caso poño float para permitir decimais
altura = float(input("Introduce a túa altura en metros: "))

# Calcular IMC usando walrus
if (imc := peso / altura**2):
    print(f"O teu IMC é: {imc:.2f}")

    # Clasificación OMS
    if imc < 18.5:
        categoria = "Baixo peso"
    elif 18.5 <= imc <= 24.9:
        categoria = "Normal"
    elif 25 <= imc <= 29.9:
        categoria = "Sobrepeso"
    else:  # imc >= 30
        categoria = "Obesidade"

    print(f"Categoria segundo OMS: {categoria}")
```

### Exercicio 10: Analizador de Texto

Analiza un texto introducido polo usuario:
1. Lonxitude.
2. Se contén certas palabras.
3. Conta vogais.
4. Verifica se é palíndromo.
5. Verifica se texto é `None`.
6. Concatena.

```python
# Pon a túa solución aquí
# Pedir texto ao usuario
texto = input("Introduce un texto: ")

# Comprobar se é None ou vacío
if texto is None or texto.strip() == "":
    print("O texto non é válido (None ou baleiro).")
else:
    # Lonxitude
    lonxitude = len(texto)
    print(f"Lonxitude do texto: {lonxitude}")

    # Contén certas palabras (por exemplo: 'python', 'curso')
    palabras_buscadas = ["python", "curso"]
    for palabra in palabras_buscadas:
        print(f"Contén '{palabra}'? {palabra.lower() in texto.lower()}")

    # Contar vogais
    vogais = "aeiouAEIOU"
    vogais_encontradas = [c for c in texto if c in vogais]
    print(f"Número total de vogais: {len(vogais_encontradas)}")
    print(f"Vogais encontradas: {''.join(vogais_encontradas)}")

    # Verificar se é palíndromo (ignora espazos e maiúsculas)
    texto_limpo = "".join(c.lower() for c in texto if c.isalnum())
    es_palindromo = texto_limpo == texto_limpo[::-1]
    print(f"É palíndromo? {es_palindromo}")

    # Concatenar con outra cadea
    texto_extra = " - Analizado co noso programa."
    texto_final = texto + texto_extra
    print("Texto concatenado:", texto_final)
```

### Xogo de Adiviñanza

Xogo onde:
- Número secreto aleatorio (1-100).
- Aportar Pistas ("é maior", "é menor").
- Contador de intentos.
- Utiliza Walrus.
- Preguntar se se quere xogar outra vez.

```python
# Pon a túa solución aquí
import random

print("Benvido ao Xogo de Adiviñanza!")

while True:
    # Número secreto aleatorio
    secreto = random.randint(1, 100)
    intentos = 0

    print("\nEstou a pensar nun número entre 1 e 100. Tenta adiviñalo!")

    while True:
        # Pedimos intento e gardamos como int usando walrus
        if (aposta := input("Introduce un número: ")):
            aposta = int(aposta)
            intentos += 1

            if aposta < secreto:
                print("É maior!")
            elif aposta > secreto:
                print("É menor!")
            else:
                print(f"Parabéns! Adiviñaches o número {secreto} en {intentos} intentos.")
                break

    # Preguntar se quere xogar outra vez
    repetir = input("Queres xogar outra vez? (si/non): ").lower()
    if repetir != "si":
        print("Gracias por xogar! Ata pronto.")
        break
```

### Validador de Email

Valida email con:
- Que conteña unha `@`.
- Que conteña un `.` despois da `@`.
- Lonxitude >= 5.
- Non conteña espazos en branco.
- Dominio pertenza a unha lista de dominios válida.

```python
# Pon a túa solución aquí
# Lista de dominios válidos
dominios_validos = ["gmail.com", "hotmail.com", "yahoo.com", "outlook.com", "edu.com", "xunta.es"]

# Pedir email ao usuario
email = input("Introduce o teu email: ").strip()  # eliminamos espazos iniciais e finais

# Función de validación
def validar_email(email):
    # 1️⃣ Lonxitude mínima
    if len(email) < 5:
        return False, "O email é demasiado curto."

    # 2️⃣ Non pode conter espazos
    if " " in email:
        return False, "O email non pode conter espazos."

    # 3️⃣ Debe conter @
    if "@" not in email:
        return False, "O email debe conter @."

    # 4️⃣ Debe conter . despois de @
    usuario, _, dominio_completo = email.partition("@")  # separa en usuario e dominio
    if "." not in dominio_completo:
        return False, "O dominio debe conter un punto."

    # 5️⃣ Dominio válido
    if dominio_completo not in dominios_validos:
        return False, f"Dominio non válido. Dominios permitidos: {', '.join(dominios_validos)}"

    # Se pasou todas as comprobacións
    return True, "Email válido!"

# Validar e mostrar resultado
valido, mensaxe = validar_email(email)
print(mensaxe)
```
